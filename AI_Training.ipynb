```
data=pd.read_csv("/content/finance_dataset.csv.crdownload")
```

```
data.head()
```

```
print(data['loan_amount'].max())
print(data['annual_income'].mean())
print(data['loan_default'].value_counts())
```

```
plt.figure(figsize=(5,5))
sns.histplot(data['annual_income'],bins=30)
plt.title("Distribution of Income")
plt.xlabel("Ammual Income")
plt.ylabel("Number of Customers")
plt.show()
```

```
plt.figure(figsize=(5,5))
sns.histplot(data['credit_score'],bins=30)
plt.title("Distribution of Credit Score")
plt.xlabel("Credit Score")
plt.ylabel("Number of Customers")
plt.show()
```

```
plt.figure(figsize=(5,5))
sns.scatterplot(x=data['credit_score'],y=data['loan_amount'])
plt.title("Credit vs Loan amount")
plt.xlabel("Credit Score")
plt.ylabel("Loan Amount")
plt.show()
```
